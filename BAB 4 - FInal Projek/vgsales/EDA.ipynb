## Data Visualization of Video Games Sales

```python
# lib data manipulation
import numpy as np
import pandas as pd

# lib data visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset
dataset = pd.read_csv("../dataset/vgsales.csv")
dataset
```

### Sum and percentage of video games sales

```python
def aggregate_sales(dataset):

  # calculate aggregate
  df = dataset[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].aggregate("sum").reset_index()

  # rename columns
  df = df.rename(columns={"index": "Region", 0: "Sales"})

  # sorting values
  df.sort_values(by="Sales", ascending=True)

  # return values
  return df
```

```python
# create figure
fig, ax = plt.subplots(figsize=(8,4))

# create barplot
sns.barplot(aggregate_sales(dataset), y="Region", x="Sales", hue="Region")

# set labels
ax.set_title("", fontsize=14)
ax.set_xlabel("", fontsize=12)
ax.set_ylabel("", fontsize=12)
ax.grid(True)

# show plot
plt.tight_layout()
plt.show()
```

### Calculate sum and percentage of Games Name, Ganre, Platform and Publisher.

```python
def grouping_sales(dataset, group):

  # process grouping by games, genre, platform, publisher.
  df = dataset.groupby(group)[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]].aggregate("sum")

  # process sorting values 
  df = df.sort_values(by=["Global_Sales"], ascending=False).reset_index().head(3)

  # return values
  return df
```
