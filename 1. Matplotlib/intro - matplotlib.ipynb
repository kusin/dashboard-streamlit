```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,4))
plt.tight_layout()
plt.show()
```

```python
# create figure
fig, ax = plt.subplots(figsize=(8,4))

# set labels
ax.set_title("This is Title of Figure", fontsize=12)
ax.set_xlabel("X Labels", fontsize=10)
ax.set_ylabel("Y Labels", fontsize=10)
ax.grid(True)

# show plot
plt.tight_layout()
plt.show()
```
